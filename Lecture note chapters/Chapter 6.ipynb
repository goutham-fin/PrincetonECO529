# Log Utility

## 1
Use goods market clearing and optimal investment to express $q^K, q^{\mathcal{B}}$ and $\iota$ in terms of $\vartheta:=\frac{q^{\mathcal{B}}}{q^K+q^{\mathcal{B}}}$

Market Clearing condition:
$$
\rho\left(q_t^{\mathcal{B}}+q_t^K\right) K_t=\left(a-\iota_t\right) K_t
$$

Optimal Investment condition:
$$
\iota_t=\frac{1}{\phi}\left(q_t^K-1\right)
$$


$$
\begin{cases}
q_t^{\mathcal{B}} & =\vartheta_t \frac{1+\phi a}{\left(1-\vartheta_t\right)+\phi\rho} \\
q_t^K & =\left(1-\vartheta_t\right) \frac{1+\phi a}{\left(1-\vartheta_t\right)+\phi \rho} \\
\iota_t & =\frac{\left(1-\vartheta_t\right) a-\rho}{\left(1-\vartheta_t\right)+\phi \rho}
\end{cases}
$$

## 2

$$
\begin{align*}
d r_t^{K, \tilde{i}} &= \left(\frac{a-\iota_t}{q_t^K} + \mu_t^{q^K} + \Phi(\iota_t) - \delta \right)dt + \sigma_t^{q^K}dZ_t + \tilde{\sigma}_td\tilde{Z}_t^{\tilde{i}}\\

d r_t^{\mathcal{B}} &= \left(\mu_t^{q^{\mathcal{B}}} + \Phi(\iota_t) - \delta \right)dt + \sigma_t^{q^{\mathcal{B}}}dZ_t \\
\end{align*}
$$

$$
\begin{align*}
\tilde{\varsigma}_t &= \tilde{\sigma}_t^{n^{\check{i}}} = \left(1-\vartheta_t\right)\tilde{\sigma}_t\\
\varsigma_t &= \sigma_t^{n^{\check{i}}} = \vartheta_t\sigma_t^{q^{\mathcal{B}}}+\left(1-\vartheta_t\right)\sigma_t^{q^K}\\
\end{align*}
$$ 

$$
\begin{align*}
\frac{\mathbb{E}\left[d r_t^{K, \tilde{i}}\right]}{d t}-\frac{\mathbb{E}\left[d r_t^{\mathcal{B}}\right]}{d t}&=\varsigma_t\left(\sigma_t^{r^{K, \bar{i}}}-\sigma_t^{r^{\mathcal{B}}}\right)+\tilde{\varsigma}_t\tilde{\sigma}_t^{K, \bar{i}}\\
\frac{a-\iota_t}{q_t^K} + \mu_t^{q^K} - \mu_t^{q^{\mathcal{B}}} &= \varsigma_t\left(\sigma_t^{r^{K, \bar{i}}}-\sigma_t^{r^{\mathcal{B}}}\right)+\tilde{\varsigma}_t\tilde{\sigma}_t^{K, \bar{i}}\\
\frac{\rho}{1-\vartheta}  + \mu_t^{q^K} - \mu_t^{q^{\mathcal{B}}} &= \varsigma_t\left(\sigma_t^{q^K} - \sigma_t^{q^{\mathcal{B}}}\right) + \tilde{\varsigma}_t\tilde{\sigma}_t
\end{align*} 
$$

By Ito's lemma,
$$
\begin{align*}
\mu_t^{\vartheta} 
&= \mu_t^{q^{\mathcal{B}}} - \mu_t^{q^K+q^{\mathcal{B}}} + \sigma_t^{q^K+q^{\mathcal{B}}}(\sigma_t^{q^K+q^{\mathcal{B}}} - \sigma_t^{q^{\mathcal{B}}})\\
&= \mu_t^{q^{\mathcal{B}}} - (1-\vartheta)\mu_t^{q^K}-\vartheta\mu_t^{q^{\mathcal{B}}} + \left[(1-\vartheta)\sigma_t^{q^K}+\vartheta\sigma_t^{q^{\mathcal{B}}}\right]\left[(1-\vartheta)\sigma_t^{q^K}+\vartheta\sigma_t^{q^{\mathcal{B}}} - \sigma_t^{q^{\mathcal{B}}}\right]\\
&= (1-\vartheta)(\mu_t^{q^{\mathcal{B}}}-\mu_t^{q^K})+ (1-\vartheta)\left[(1-\vartheta)\sigma_t^{q^K}+\vartheta\sigma_t^{q^{\mathcal{B}}}\right]\left(\sigma_t^{q^K}- \sigma_t^{q^{\mathcal{B}}}\right)\\
& \text{plug in the martingale pricing condition}\\
\mu_t^{\vartheta}
&= \rho - (1-\vartheta)^2(\tilde{\sigma}_t)^2
\end{align*}
$$

## 4
$$
d \tilde{\sigma}_t=b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) d t+\nu \sqrt{\tilde{\sigma}_t} d Z_t
$$
$$
\begin{align*}
d\vartheta(\tilde{\sigma}_t) 
&= \vartheta'(\tilde{\sigma}_t)d\tilde{\sigma}_t + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)(d\tilde{\sigma}_t)^2\\
&= \vartheta'(\tilde{\sigma}_t)\left[b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) d t+\nu \sqrt{\tilde{\sigma}_t} d Z_t\right] + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_tdt\\
\vartheta_t\mu_t^{\vartheta} &= \vartheta'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t
\end{align*}
$$

$$
\vartheta(\rho - (1-\vartheta)^2(\tilde{\sigma}_t)^2) = \vartheta'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t
$$

$$
\rho\vartheta = \vartheta(1-\vartheta)^2(\tilde{\sigma}_t)^2 + \vartheta'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t
$$

```python
batch_size = 100 # number of grids for state variable

# define the PDE model
pde_model_log = PDEModel(name = "Monetary Theory",              # name of the model, used only for display
                     config = {"sampling_method": SamplingMethod.UniformRandom,   # sampling method for the state variable
                               "batch_size"    : batch_size, # number of grids for the state variable
                               "num_epochs"    : 2000,      # number of epochs for training
                               "optimizer_type": OptimizerType.Adam})         # optimizer type / algorithm

# Step 1: define the state variables
# we define the state variable sigma_tilde and its range by Gamma distribution
pde_model_log.set_state(["sigmatilde"], {"sigmatilde": [sigma_min, sigma_max]})  # define the state variable and its range

# Step 2: define learnable variables. vartheta is the only learnable variable in this model
# configs includes the structure of each NN. We use ActivationType.SiLU (check detail here) as the activation function.
pde_model_log.add_endogs(names = ["vartheta"], # endogenous variables in a list
                     configs={"vartheta": {"positive": True,  "activation_type": ActivationType.SiLU}})

# Step 3: define parameters.
pde_model_log.add_params(params) 

# Step 4: add the boundary conditions
# no boundary conditions for this model

# Step 5: add equations for learnable variables. 
# pde_model.add_endog_equation(r"$\rho*\vartheta = \vartheta*(1-\vartheta)^2*(\tilde{\sigma})^2 + \frac{\partial\vartheta}{\partial\tilde{\sigma}}*b*\left(\sigma_{ss}-\tilde{\sigma}\right) + \frac{1}{2}*\frac{\partial^2\vartheta}{\partial\tilde{\sigma}^2}*\nu^2*\tilde{\sigma}$", weight = batch_size)
pde_model_log.add_endog_equation("rho*vartheta = vartheta*(1-vartheta)**2*(sigmatilde)**2 + vartheta_sigmatilde*b*(sigmass-sigmatilde) + 0.5*vartheta_sigmatildesigmatilde*nu**2*sigmatilde", weight = batch_size)

# Step 6: add equations for other variables
# no other equations for this model

pde_model_log.train_model("./models/log_u", "log_u.pt", True)     # train and save the model
pde_model_log.eval_model(True)                                    # evaluate the model: print the losses and variable list    
```

In order to calculate the risk free rate, we need to first calculate the price of risk $\tilde{\varsigma}_t$ and $\varsigma_t$.
$$
\begin{align*}
\tilde{\varsigma}_t &= \tilde{\sigma}_t^{n^{\check{i}}} = \left(1-\theta_t^{\tilde{i}}\right)\tilde{\sigma}_t\\
\varsigma_t &= \sigma_t^{n^{\check{i}}} = \theta_t^{\tilde{i}}\sigma_t^{q^{\mathcal{B}}}+\left(1-\theta_t^{\tilde{i}}\right)\sigma_t^{q^K}\\
\end{align*}
$$ 

By the asset pricing equations, we have the below two equivalent equations.
$$
\begin{cases}
r^f = (a-\iota_t)/q^K + \Phi(\iota_t) - \delta + \mu_t^{q^K} - \varsigma_t\sigma_t^{q^K} - \tilde{\varsigma}_t\tilde{\sigma}_t\\
r^f = \mu_t^{q^{\mathcal{B}}} + \Phi(\iota_t) - \delta - \varsigma_t\sigma_t^{q^{\mathcal{B}}}
\end{cases}
$$
Therefore, we need to calculate $\mu_t^{q^{\mathcal{B}}}$, $\mu_t^{q^K}$, $\sigma_t^{q^{\mathcal{B}}}$, $\sigma_t^{q^K}$


$$
\begin{align*}
dq^{\mathcal{B}} = q^{\mathcal{B}'}(\vartheta)d\vartheta 
&= q^{\mathcal{B}'}(\vartheta)\left(\vartheta'(\tilde{\sigma})d\tilde{\sigma} + \frac{1}{2}\vartheta''(\tilde{\sigma})(d\tilde{\sigma})^2\right)\\
dq^{K} = q^{K'}(\vartheta)d\vartheta 
&= q^{K'}(\vartheta)\left(\vartheta'(\tilde{\sigma})d\tilde{\sigma} + \frac{1}{2}\vartheta''(\tilde{\sigma})(d\tilde{\sigma})^2\right)\\
\end{align*}
$$
$$
d \tilde{\sigma}_t=b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) d t+\nu \sqrt{\tilde{\sigma}_t} d Z_t,
$$
$$
\mu^{q^{\mathcal{B}}} = \frac{q^{\mathcal{B}'}(\vartheta)}{q^{\mathcal{B}}(\vartheta)}\left[\vartheta'(\tilde{\sigma})b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) +  \frac{1}{2}\vartheta''(\tilde{\sigma})\nu^2 \tilde{\sigma}_t \right]
$$
$$
\mu^{q^K} = \frac{q^{K'}(\vartheta)}{q^{K}(\vartheta)}\left[\vartheta'(\tilde{\sigma})b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) +  \frac{1}{2}\vartheta''(\tilde{\sigma})\nu^2 \tilde{\sigma}_t \right]
$$
$$
\sigma^{q^{\mathcal{B}}} = \frac{q^{{\mathcal{B}}'}(\vartheta)}{q^{{\mathcal{B}}}(\vartheta)}\vartheta'(\tilde{\sigma})\nu \sqrt{\tilde{\sigma}_t}
$$
$$
\sigma^{K} = \frac{q^{K'}(\vartheta)}{q^{K}(\vartheta)}\vartheta'(\tilde{\sigma})\nu \sqrt{\tilde{\sigma}_t}
$$

```python
sigmatilde                    = torch.linspace(sigma_min, sigma_max, 1000, device=device).reshape(-1,1)
vartheta                      = pde_model_log.endog_vars["vartheta"](sigmatilde).detach().cpu().numpy()
vartheta_sigmatilde           = pde_model_log.local_function_dict['vartheta_sigmatilde'](sigmatilde).detach().cpu().numpy()
vartheta_sigmatildesigmatilde = pde_model_log.local_function_dict['vartheta_sigmatildesigmatilde'](sigmatilde).detach().cpu().numpy()
sigmatilde                    = sigmatilde.detach().cpu().numpy()

iota          = ((1-vartheta)*a - rho)/((1-vartheta) + phi*rho)
qB            = vartheta*(1+phi*a)/((1-vartheta) + phi*rho)
qK            = (1-vartheta)*(1+phi*a)/((1-vartheta) + phi*rho)
qB_sigmatilde = (1+phi*a)/((1-vartheta) + phi*rho) + (1+phi*a)*vartheta/((1-vartheta) + phi*rho)**2
qK_sigmatilde = -(1+phi*a)/((1-vartheta) + phi*rho) + (1+phi*a)*(1-vartheta)/((1-vartheta) + phi*rho)**2

sigmaqB       = qB_sigmatilde/qB*vartheta_sigmatilde*nu*sigmatilde**0.5
sigmaqK       = qK_sigmatilde/qK*vartheta_sigmatilde*nu*sigmatilde**0.5
muqB          = qB_sigmatilde/qB*(vartheta_sigmatilde*b*(sigmass-sigmatilde) + 0.5*vartheta_sigmatildesigmatilde*sigmatilde*nu**2)
muqK          = qK_sigmatilde/qK*(vartheta_sigmatilde*b*(sigmass-sigmatilde) + 0.5*vartheta_sigmatildesigmatilde*sigmatilde*nu**2)
varsigma      = vartheta*sigmaqB + (1-vartheta)*sigmaqK
varsigmatilde = (1-vartheta)*sigmatilde


rf_qB = muqB + np.log(phi*iota+1)/phi - delta - varsigma * sigmaqB
rf_qK = (a-iota)/qK + np.log(phi*iota+1)/phi - delta + muqK - varsigma * sigmaqK - varsigmatilde*sigmatilde


fig, ax = plt.subplots(2,3, figsize=(15,10))
ax[0,0].plot(sigmatilde, vartheta)
ax[0,0].set_ylabel(r'$\vartheta$')
ax[0,0].set_xlabel(r'$\tilde{\sigma}$')
ax[0,1].plot(sigmatilde, qK)
ax[0,1].set_ylabel(r'$q^K$')
ax[0,1].set_xlabel(r'$\tilde{\sigma}$')
ax[0,2].plot(sigmatilde, qB)
ax[0,2].set_ylabel(r'$q^B$')
ax[0,2].set_xlabel(r'$\tilde{\sigma}$')
ax[1,0].plot(sigmatilde, rf_qK, label=r'$r^{f}$ calculated by $q^K$')
ax[1,0].plot(sigmatilde, rf_qB, label=r'$r^{f}$ calculated by $q^\mathcal{B}$')
ax[1,0].set_ylabel(r'$r^{f}$')
ax[1,0].set_xlabel(r'$\tilde{\sigma}$')
ax[1,0].legend()
ax[1,1].plot(sigmatilde, varsigma)
ax[1,1].set_ylabel(r'$\varsigma$')
ax[1,1].set_xlabel(r'$\tilde{\sigma}$')
ax[1,2].plot(sigmatilde, varsigmatilde)
ax[1,2].set_ylabel(r'$\tilde{\varsigma}$')
ax[1,2].set_xlabel(r'$\tilde{\sigma}$')
plt.tight_layout()
plt.show()
```

# CRRA Utility

## 1

Question 1 is the same as the log utility case, except that the optimal consumption-wealth ratio is different.

$$
\begin{cases}
q_t^{\mathcal{B}} & =\vartheta_t \frac{1+\phi a}{\left(1-\vartheta_t\right)+\phi\check{\rho}} \\
q_t^K & =\left(1-\vartheta_t\right) \frac{1+\phi a}{\left(1-\vartheta_t\right)+\phi \check{\rho}} \\
\iota_t & =\frac{\left(1-\vartheta_t\right) a-\check{\rho}}{\left(1-\vartheta_t\right)+\phi \check{\rho}}
\end{cases}
$$

## 2

Postulate that 
$$
V_t^{\tilde{i}} =  \frac{(v_t\tilde{\eta}_tK_t)^{1-\gamma}}{1-\gamma}
$$

By the optimal consumption condition, we have $\frac{\partial V_t^{\tilde{i}}}{\partial n_t^{\tilde{i}}} = u'(c_t^{\tilde{i}})$. 
$$
(c_t^{\tilde{i}})^{-\gamma} =  v_t^{1-\gamma} \tilde{\eta}_t^{-\gamma} K_t^{-\gamma} (\frac{1}{q_t^{\mathcal{B}}+q_t^K})
$$
Therefore, 
$$
\check{\rho}^{-\gamma} =  \left(\frac{v_t}{q^{\mathcal{B}}_t + q^{K}_t}\right)^{1-\gamma}
$$

$$
-\gamma\sigma^{c^{\tilde{i}}} = (1-\gamma)\sigma_t^{v} - \gamma\sigma_t^{\tilde{\eta}}  - \sigma_t^{q^{\mathcal{B}}+q^K}
$$

The sum of the price of risk is $\tilde{\varsigma}_t + \varsigma_t =  \gamma\sigma^{c^{\tilde{i}}}$, where $\tilde{\varsigma}_t$ is the price of idiosyncratic risk and $\varsigma_t$ is the price of systematic risk. Therefore, the price of risk $\tilde{\varsigma}_t$ and $\varsigma_t$ are 
$$
\begin{align*}
\tilde{\varsigma}_t &= \gamma\tilde{\sigma}_t^{n^{\check{i}}} = \gamma\left(1-\vartheta_t\right)\tilde{\sigma}_t\\
\varsigma_t &= (\gamma-1)\sigma_t^{v} + \sigma_t^{q^{\mathcal{B}}+q^K} = (\gamma-1)\sigma_t^{v} + \vartheta_t\sigma_t^{q^{\mathcal{B}}}+\left(1-\vartheta_t\right)\sigma_t^{q^K}\\
\end{align*}
$$ 
where we based on the evolution of $\tilde{\eta}_t^{\tilde{i}}$
$$
\frac{\mathrm{d} \tilde{\eta}_t^{\tilde{i}}}{\tilde{\eta}_t^{\tilde{i}}}=\frac{\mathrm{d}\left(n_t^{\tilde{i}} / N_t^i\right)}{n_t^{\tilde{i}} / N_t^i}=\left(\mu_t^{n^{\tilde{i}}}-\mu_t^N+\left(\sigma_t^N\right)^2-\sigma^N \sigma_t^{n^{\tilde{i}}}\right) \mathrm{d} t+\left(\sigma_t^{n^{\tilde{i}}}-\sigma^N\right) \mathrm{d} Z_t+\tilde{\sigma}_t^{n^{\tilde{i}}} \mathrm{~d} \tilde{Z}_t^{\tilde{i}}=\tilde{\sigma}_t^{n^{\tilde{i}}} \mathrm{~d} \tilde{Z}_t^{\tilde{i}}
$$



The below is similar to the log utility case. By the two asset pricing equations, we have the martingale pricing condition
$$
\frac{\check{\rho}}{1-\vartheta}  + \mu_t^{q^K} - \mu_t^{q^{\mathcal{B}}} = \varsigma_t\left(\sigma_t^{q^K} - \sigma_t^{q^{\mathcal{B}}}\right) + \tilde{\varsigma}_t\tilde{\sigma}_t
$$


By Ito's lemma,
$$
\begin{align*}
\mu_t^{\vartheta} 
&= \mu_t^{q^{\mathcal{B}}} - \mu_t^{q^K+q^{\mathcal{B}}} + \sigma_t^{q^K+q^{\mathcal{B}}}(\sigma_t^{q^K+q^{\mathcal{B}}} - \sigma_t^{q^{\mathcal{B}}})\\
&= \mu_t^{q^{\mathcal{B}}} - (1-\vartheta)\mu_t^{q^K}-\vartheta\mu_t^{q^{\mathcal{B}}} + \left[(1-\vartheta)\sigma_t^{q^K}+\vartheta\sigma_t^{q^{\mathcal{B}}}\right]\left[(1-\vartheta)\sigma_t^{q^K}+\vartheta\sigma_t^{q^{\mathcal{B}}} - \sigma_t^{q^{\mathcal{B}}}\right]\\
&= (1-\vartheta)(\mu_t^{q^{\mathcal{B}}}-\mu_t^{q^K})+ (1-\vartheta)\left[(1-\vartheta)\sigma_t^{q^K}+\vartheta\sigma_t^{q^{\mathcal{B}}}\right]\left(\sigma_t^{q^K}- \sigma_t^{q^{\mathcal{B}}}\right)\\
& \text{plug in the martingale pricing condition}\\
\mu_t^{\vartheta}
&= \check{\rho} - \gamma(1-\vartheta)^2(\tilde{\sigma}_t)^2 + (1-\vartheta)\sigma_t^{v}(\sigma^{q^K} - \sigma^{q^{\mathcal{B}}})
\end{align*}
$$

Recall that $q^K/q^{\mathcal{B}} = \vartheta/(1-\vartheta)$. Therefore, 
$$
\sigma^{q^K} - \sigma^{q^{\mathcal{B}}} = \sigma^\vartheta - \sigma^{(1-\vartheta)} = \sigma^\vartheta + \frac{\vartheta}{1-\vartheta}\sigma^\vartheta
$$

Therefore, 
$$
\mu_t^{\vartheta} = \check{\rho} - \gamma(1-\vartheta)^2(\tilde{\sigma}_t)^2 + \sigma_t^{v}\sigma^{\vartheta}
$$

## 4

The difference between CRRA and log utility is that we need to consider the evolution of $v_t$, i.e. we need to know $\sigma^{\vartheta}$, so we can calculate $\vartheta$ in the above equation.

By the optimal consumption condition, we know that $\frac{dV_t^{\tilde{i}}}{V_t^{\tilde{i}}} - \rho dt + \check{\rho}dt$ is a martingale, where

$$
\frac{dV_t^{\tilde{i}}}{V_t^{\tilde{i}}} = \frac{d\left[(v_t\tilde{\eta}_tK_t)^{1-\gamma}\right]}{(v_t\tilde{\eta}_tK_t)^{1-\gamma}} = \frac{d(v_t^{1-\gamma})}{v_t^{1-\gamma}} + \frac{d\tilde{\eta}_t^{1-\gamma}}{\tilde{\eta}_t^{1-\gamma}} + \frac{dK_t^{1-\gamma}}{K_t^{1-\gamma}}
$$

We can ignore the interaction terms in the above equation, because the diffusion terms of $(v_t, \tilde{\eta}_t, K_t)$ are independent. Specifically, the diffusion term of $v_t$ is $dZ_t$; the diffusion term of $\tilde{\eta}_t$ is $d\tilde{Z}_t^{\tilde{i}}$; the diffusion term of $K_t$ is zero.

Therefore, the BSDE for $v_t$ is

$$
(1-\gamma)\left(\mu_t^v+\boldsymbol{\Phi}\left(\iota_t\right)-\delta\right)-\frac{1}{2} \gamma(1-\gamma)\left[(\sigma^{v})^2+\left(\tilde{\sigma}^{n^i}\right)^2\right]=\rho-\frac{c_t^{\tilde{i}}}{n_t^{\tilde{i}}}
$$

$$
(1-\gamma)\left(\mu_t^v+\frac{1}{\phi}log\left[\phi\iota + 1\right]-\delta\right)-\frac{1}{2} \gamma(1-\gamma)\left[(\sigma^{v})^2+\left(\left(1-\vartheta_t\right)\tilde{\sigma}_t\right)^2\right]=\rho-\check{\rho}
$$


Similar to the log utility case, we have the following equations for $\mu^{\vartheta}$ and $\sigma^{\vartheta}$.
$$
d \tilde{\sigma}_t=b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) d t+\nu \sqrt{\tilde{\sigma}_t} d Z_t
$$
$$
\begin{align*}
d\vartheta(\tilde{\sigma}_t) 
&= \vartheta'(\tilde{\sigma}_t)d\tilde{\sigma}_t + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)(d\tilde{\sigma}_t)^2\\
&= \vartheta'(\tilde{\sigma}_t)\left[b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) d t+\nu \sqrt{\tilde{\sigma}_t} d Z_t\right] + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_tdt\\
\vartheta_t\mu_t^{\vartheta} &= \vartheta'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t \\
\vartheta_t\sigma_t^{v} &= \vartheta_t'(\tilde{\sigma}_t)\nu\sqrt{\tilde{\sigma}_t}
\end{align*}
$$


In addition to $\vartheta$, we also have the following equations for $\mu^v$ and $\sigma^v$ in the CRRA case.
$$
\begin{align*}
dv(\tilde{\sigma}_t) 
&= v'(\tilde{\sigma}_t)d\tilde{\sigma}_t + \frac{1}{2}v''(\tilde{\sigma}_t)(d\tilde{\sigma}_t)^2\\
&= v'(\tilde{\sigma}_t)\left[b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) d t+\nu \sqrt{\tilde{\sigma}_t} d Z_t\right] + \frac{1}{2}v''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_tdt\\
v_t\mu_t^{v} &= v'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) + \frac{1}{2}v''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t\\
v_t\sigma_t^{v} &= v'(\tilde{\sigma}_t)\nu\sqrt{\tilde{\sigma}_t}
\end{align*}
$$

## Code implementation

Endogenous equations:
$$
(1-\gamma)\left[\mu_t^v+\frac{1}{\phi}log\left[\phi\iota + 1\right]-\delta\right]-\frac{1}{2} \gamma(1-\gamma)\left[(\sigma^{v})^2+\left(\left(1-\vartheta_t\right)\tilde{\sigma}_t\right)^2\right] - \rho + \check{\rho} = 0
$$
$$
\check{\rho}^{(\frac{\gamma}{\gamma-1})} - \left(\frac{v_t}{q^{\mathcal{B}}_t + q^{K}_t}\right) = 0
$$
$$
\iota_t - \frac{\left(1-\vartheta_t\right) a-\check{\rho}}{\left(1-\vartheta_t\right)+\phi \check{\rho}} = 0
$$
$$
\mu_t^{\vartheta} -  \check{\rho} + \gamma(1-\vartheta)^2(\tilde{\sigma}_t)^2 - \sigma_t^{v}\sigma^{\vartheta} = 0
$$

Equations:
$$
q_t^{\mathcal{B}}  =\vartheta_t \frac{1+\phi a}{\left(1-\vartheta_t\right)+\phi\check{\rho}}
$$
$$
q_t^K  =\left(1-\vartheta_t\right) \frac{1+\phi a}{\left(1-\vartheta_t\right)+\phi \check{\rho}}
$$
$$
\mu_t^{\vartheta} = \frac{1}{\vartheta_t}\left[\vartheta'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) + \frac{1}{2}\vartheta''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t \right]
$$
$$
\sigma_t^{v} = \frac{1}{\vartheta_t}\left[\vartheta_t'(\tilde{\sigma}_t)\nu\sqrt{\tilde{\sigma}_t}\right]

\mu_t^{v} = \frac{1}{v_t}\left[v'(\tilde{\sigma}_t)b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right) + \frac{1}{2}v''(\tilde{\sigma}_t)\nu^2\tilde{\sigma}_t\right]
$$
$$
\sigma_t^{v} = \frac{1}{v_t}\left[v'(\tilde{\sigma}_t)\nu\sqrt{\tilde{\sigma}_t}\right]
$$

```python
params = {
    "a"      : 0.2,
    "phi"    : 1,
    "delta"  : 0.05,
    "rho"    : 0.01,
    "sigmass": 0.2,
    "b"      : 0.05,
    "nu"     : 0.02,
    "gamma"  : 2
            }
globals().update(params) # unpack parameters

from scipy.stats import gamma as gamma_dist
alpha = 2*b*sigmass/nu**2
beta = 2*b/nu**2
scale = 1 / beta
sigma_min, sigma_max = gamma_dist.interval(0.9999999999,a=alpha, loc=0, scale=1/beta)
```

```python
batch_size = 100 # number of grids for state variable
# define the PDE model
pde_model = PDEModel(name = "Monetary Theory",              # name of the model, used only for display
                     config = {"sampling_method": SamplingMethod.UniformRandom,   # sampling method for the state variable
                               "batch_size"    : batch_size, # number of grids for the state variable
                               "num_epochs"    : 10000,      # number of epochs for training
                               "optimizer_type": OptimizerType.Adam, # optimizer type / algorithm
                               "loss_log_interval": 10,
                               "lr": 0.001
                               })    

# Step 1: define the state variables
# we define the state variable sigma_tilde and its range by Gamma distribution
pde_model.set_state(["sigmatilde"], {"sigmatilde": [sigma_min, sigma_max]})  # define the state variable and its range

# Step 2: define learnable variables. vartheta is the only learnable variable in this model
# configs includes the structure of each NN. We use ActivationType.SiLU (check detail here) as the activation function.
pde_model.add_endogs(names = ["vartheta", "v","iota", "rhocheck"], # endogenous variables in a list
                     configs={"vartheta": {"positive": True,  "activation_type": ActivationType.SiLU},
                              "v" : {"positive": True, "activation_type": ActivationType.SiLU},
                              "iota": {"positive": True, "activation_type": ActivationType.SiLU},
                              "rhocheck": {"positive": True, "activation_type": ActivationType.SiLU}})

# Step 3: define parameters.
pde_model.add_params(params) 

# Step 4: add the boundary conditions
# no boundary conditions for this model

# Step 5: add equations for learnable variables. 
pde_model.add_endog_equation("(1-gamma)*(muv + 1/phi*log(phi*iota + 1)-delta) - 0.5*gamma*(1-gamma)*(sigmav**2 + ((1-vartheta)*sigmatilde)**2) - rho + rhocheck = 0")
pde_model.add_endog_equation("rhocheck**(gamma/(gamma-1)) - (v/(qB+qK)) = 0")
pde_model.add_endog_equation("iota - ((1-vartheta)*a - rhocheck)/((1-vartheta) + phi*rhocheck) = 0")
pde_model.add_endog_equation("muvartheta - rhocheck + gamma*(1-vartheta)**2 * sigmatilde**2 - sigmav*sigmavartheta = 0")

# Step 6: add equations for other variables

pde_model.add_equation("qB = vartheta*(1+phi*a)/((1-vartheta) + phi*rhocheck)")
pde_model.add_equation("qK = (1-vartheta)*(1+phi*a)/((1-vartheta) + phi*rhocheck)")
pde_model.add_equation("muvartheta = 1/vartheta*(vartheta_sigmatilde*b*(sigmass-sigmatilde) + 0.5 * vartheta_sigmatildesigmatilde * nu**2 * sigmatilde)")
pde_model.add_equation("sigmavartheta = 1/vartheta*vartheta_sigmatilde*nu*sigmatilde**0.5")
pde_model.add_equation("muv = 1/v*(v_sigmatilde*b*(sigmass-sigmatilde) + 0.5*v_sigmatildesigmatilde*nu**2*sigmatilde)")
pde_model.add_equation("sigmav = 1/v*v_sigmatilde*nu*sigmatilde**0.5")

pde_model.train_model("./models/crra_u", "crra_u.pt", True)     # train and save the model
pde_model.eval_model(True)                                    # evaluate the model: print the losses and variable list
```

In order to calculate the risk free rate, we need to first calculate the price of risk $\tilde{\varsigma}_t$ and $\varsigma_t$.
$$
\begin{align*}
\tilde{\varsigma}_t &= \gamma\tilde{\sigma}_t^{n^{\check{i}}} = \gamma\left(1-\vartheta_t\right)\tilde{\sigma}_t\\
\varsigma_t &= -\sigma_t^{v} + \sigma_t^{q^{\mathcal{B}}+q^K} = -\sigma_t^{v} + \vartheta_t\sigma_t^{q^{\mathcal{B}}}+\left(1-\vartheta_t\right)\sigma_t^{q^K}\\
\end{align*}
$$ 
By the asset pricing equations, we have the below two equivalent equations.
$$
\begin{cases}
r^f = (a-\iota_t)/q^K + \Phi(\iota_t) - \delta + \mu_t^{q^K} - \varsigma_t\sigma_t^{q^K} - \tilde{\varsigma}_t\tilde{\sigma}_t\\
r^f = \mu_t^{q^{\mathcal{B}}} + \Phi(\iota_t) - \delta - \varsigma_t\sigma_t^{q^{\mathcal{B}}}
\end{cases}
$$
Therefore, we need to calculate $\mu_t^{q^{\mathcal{B}}}$, $\mu_t^{q^K}$, $\sigma_t^{q^{\mathcal{B}}}$, $\sigma_t^{q^K}$

The first method is to take derivative of $q^{\mathcal{B}}$ and $q^K$ w.r.t. $\tilde{\sigma}_t$.
$$
\begin{align*}
dq^{\mathcal{B}} &= \frac{\partial q^{\mathcal{B}}}{\partial\tilde{\sigma}}d\tilde{\sigma} + \frac{1}{2}\frac{\partial^2 q^{\mathcal{B}}}{\partial\tilde{\sigma}^2}(d\tilde{\sigma})^2\\
dq^{K} &= \frac{\partial q^{K}}{\partial\tilde{\sigma}}d\tilde{\sigma} + \frac{1}{2}\frac{\partial^2 q^{K}}{\partial\tilde{\sigma}^2}(d\tilde{\sigma})^2\\
\end{align*}
$$
$$
q^{\mathcal{B}}\mu^{q^{\mathcal{B}}} = \frac{\partial q^{\mathcal{B}}}{\partial\tilde{\sigma}}\left(b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right)\right) + \frac{1}{2}\frac{\partial^2 q^{\mathcal{B}}}{\partial\tilde{\sigma}^2}\nu^2\tilde{\sigma}_t
$$
$$
q^{\mathcal{B}}\sigma^{q^{\mathcal{B}}} = \frac{\partial q^{\mathcal{B}}}{\partial\tilde{\sigma}}\nu \sqrt{\tilde{\sigma}_t}
$$
$$
q^{K}\mu^{q^{K}} = \frac{\partial q^{K}}{\partial\tilde{\sigma}}\left(b\left(\tilde{\sigma}^{ss}-\tilde{\sigma}_t\right)\right) + \frac{1}{2}\frac{\partial^2 q^{K}}{\partial\tilde{\sigma}^2}\nu^2\tilde{\sigma}_t
$$
$$
q^{K}\sigma^{q^{K}} = \frac{\partial q^{K}}{\partial\tilde{\sigma}}\nu \sqrt{\tilde{\sigma}_t}
$$

The second method is to take derivative of $q^{\mathcal{B}}$ and $q^K$ w.r.t. $\vartheta$ and $\check{\rho}$, and then use the chain rule to calculate $\mu_t^{q^{\mathcal{B}}}$, $\mu_t^{q^K}$, $\sigma_t^{q^{\mathcal{B}}}$, $\sigma_t^{q^K}$.

$$
\begin{align*}
dq^{\mathcal{B}} &= \frac{\partial q^{\mathcal{B}}}{\partial\vartheta}d\vartheta + \frac{\partial q^{\mathcal{B}}}{\partial \check{\rho}}d\check{\rho}\\
&= \frac{\partial q^{\mathcal{B}}}{\partial\vartheta}\left(\vartheta'(\tilde{\sigma})d\tilde{\sigma} + \frac{1}{2}\vartheta''(\tilde{\sigma})(d\tilde{\sigma})^2\right) + \frac{\partial q^{\mathcal{B}}}{\partial \check{\rho}}\left(\check{\rho}'(\tilde{\sigma})d\tilde{\sigma} + \frac{1}{2}\check{\rho}''(\tilde{\sigma})(d\tilde{\sigma})^2\right) \\
dq^{K} &= \frac{\partial q^{K}}{\partial\vartheta}d\vartheta + \frac{\partial q^{K}}{\partial \check{\rho}}d\check{\rho}\\
&= \frac{\partial q^{K}}{\partial\vartheta}\left(\vartheta'(\tilde{\sigma})d\tilde{\sigma} + \frac{1}{2}\vartheta''(\tilde{\sigma})(d\tilde{\sigma})^2\right) + \frac{\partial q^{K}}{\partial \check{\rho}}\left(\check{\rho}'(\tilde{\sigma})d\tilde{\sigma} + \frac{1}{2}\check{\rho}''(\tilde{\sigma})(d\tilde{\sigma})^2\right) \\
\end{align*}
$$
$$
q^{\mathcal{B}}\mu^{q^{\mathcal{B}}} = 
\left(
    \frac{\partial q^{\mathcal{B}}}{\partial\vartheta}\vartheta'(\tilde{\sigma})
    + \frac{\partial q^{\mathcal{B}}}{\partial \check{\rho}}\check{\rho}'(\tilde{\sigma})
\right)
b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) +  
\frac{1}{2}
\left(
\frac{\partial q^{\mathcal{B}}}{\partial\vartheta}\vartheta''(\tilde{\sigma})
+ \frac{\partial q^{\mathcal{B}}}{\partial \check{\rho}}\check{\rho}''(\tilde{\sigma})
\right)
\nu^2 \tilde{\sigma}_t 
$$
$$
q^{K}\mu^{q^{K}} = 
\left(
    \frac{\partial q^{K}}{\partial\vartheta}\vartheta'(\tilde{\sigma})
    + \frac{\partial q^{K}}{\partial \check{\rho}}\check{\rho}'(\tilde{\sigma})
\right)
b\left(\tilde{\sigma}^{s s}-\tilde{\sigma}_t\right) +  
\frac{1}{2}
\left(
\frac{\partial q^{K}}{\partial\vartheta}\vartheta''(\tilde{\sigma})
+ \frac{\partial q^{K}}{\partial \check{\rho}}\check{\rho}''(\tilde{\sigma})
\right)
\nu^2 \tilde{\sigma}_t 
$$
$$
q^{\mathcal{B}}\sigma^{q^{\mathcal{B}}} = \left(
    \frac{\partial q^{\mathcal{B}}}{\partial\vartheta}\vartheta'(\tilde{\sigma})
    + \frac{\partial q^{\mathcal{B}}}{\partial \check{\rho}}\check{\rho}'(\tilde{\sigma})
\right)\nu \sqrt{\tilde{\sigma}_t}
$$
$$
q^{K}\sigma^{K} = \left(
    \frac{\partial q^{K}}{\partial\vartheta}\vartheta'(\tilde{\sigma})
    + \frac{\partial q^{K}}{\partial \check{\rho}}\check{\rho}'(\tilde{\sigma})
\right)\nu \sqrt{\tilde{\sigma}_t}
$$

```python
sigmatilde = torch.linspace(sigma_min, sigma_max, batch_size, device=device).reshape(-1,1)
vartheta   = pde_model.endog_vars["vartheta"](sigmatilde).detach().cpu().numpy()
iota       = pde_model.endog_vars["iota"](sigmatilde).detach().cpu().numpy()
rhocheck   = pde_model.endog_vars['rhocheck'](sigmatilde).detach().cpu().numpy()
v          = pde_model.endog_vars["v"](sigmatilde).detach().cpu().numpy()

vartheta_sigmatilde           = pde_model.local_function_dict['vartheta_sigmatilde'](sigmatilde).detach().cpu().numpy()
vartheta_sigmatildesigmatilde = pde_model.local_function_dict['vartheta_sigmatildesigmatilde'](sigmatilde).detach().cpu().numpy()
rhocheck_sigmatilde           = pde_model.local_function_dict['rhocheck_sigmatilde'](sigmatilde).detach().cpu().numpy()
rhocheck_sigmatildesigmatilde = pde_model.local_function_dict['rhocheck_sigmatildesigmatilde'](sigmatilde).detach().cpu().numpy()
v_sigmatilde                  = pde_model.local_function_dict['v_sigmatilde'](sigmatilde).detach().cpu().numpy()
v_sigmatildesigmatilde        = pde_model.local_function_dict['v_sigmatildesigmatilde'](sigmatilde).detach().cpu().numpy()

sigmatilde = sigmatilde.detach().cpu().numpy()

qB            = vartheta*(1+phi*a)/((1-vartheta) + phi*rhocheck)
qK            = (1-vartheta)*(1+phi*a)/((1-vartheta) + phi*rhocheck)
muvartheta    = 1/vartheta*(vartheta_sigmatilde*b*(sigmass-sigmatilde) + 0.5 * vartheta_sigmatildesigmatilde * nu**2 * sigmatilde)
sigmavartheta = 1/vartheta*vartheta_sigmatilde*nu*sigmatilde**0.5
muv           = 1/v*(v_sigmatilde*b*(sigmass-sigmatilde) + 0.5*v_sigmatildesigmatilde*nu**2*sigmatilde)
sigmav        = 1/v*v_sigmatilde*nu*sigmatilde**0.5

# method 1
################################################
qB_sigmatilde = np.gradient(qB.flatten(), sigmatilde.flatten()).reshape(-1,1)   
qK_sigmatilde = np.gradient(qK.flatten(), sigmatilde.flatten()).reshape(-1,1)
qB_sigmatildesigmatilde = np.gradient(qB_sigmatilde.flatten(), sigmatilde.flatten()).reshape(-1,1)
qK_sigmatildesigmatilde = np.gradient(qK_sigmatilde.flatten(), sigmatilde.flatten()).reshape(-1,1)

sigmaqB       = 1/qB * (qB_sigmatilde*params['nu'] * sigmatilde**0.5)
sigmaqK       = 1/qK * (qK_sigmatilde*params['nu'] * sigmatilde**0.5)
muqB          = 1/qB * (qB_sigmatilde*params['b']*(params['sigmass']-sigmatilde) + 0.5 * qB_sigmatildesigmatilde *params['nu']**2 * sigmatilde)
muqK          = 1/qK * (qK_sigmatilde*params['b']*(params['sigmass']-sigmatilde) + 0.5 * qK_sigmatildesigmatilde *params['nu']**2 * sigmatilde) 
################################################

# method 2
################################################
# qB_vartheta = (1+phi*a)/((1-vartheta) + phi*rhocheck) + (1+phi*a)*vartheta/((1-vartheta) + phi*rhocheck)**2
# qK_vartheta = -(1+phi*a)/((1-vartheta) + phi*rhocheck) + (1+phi*a)*(1-vartheta)/((1-vartheta) + phi*rhocheck)**2
# qB_rhocheck = vartheta*(1+phi*a)/(1-vartheta+phi*rhocheck)**2
# qK_rhocheck = (1-vartheta)*(1+phi*a)/(1-vartheta+phi*rhocheck)**2

# sigmaqB       = 1/qB * (qB_vartheta*vartheta_sigmatilde + qB_rhocheck*rhocheck_sigmatilde) *nu * sigmatilde**0.5
# sigmaqK       = 1/qK * (qK_vartheta*vartheta_sigmatilde + qK_rhocheck*rhocheck_sigmatilde) *nu * sigmatilde**0.5
# muqB          = 1/qB * (qB_vartheta*vartheta_sigmatilde + qB_rhocheck*rhocheck_sigmatilde) *b*(sigmass-sigmatilde) + 0.5 * (qB_vartheta*vartheta_sigmatildesigmatilde + qB_rhocheck*rhocheck_sigmatildesigmatilde) *nu**2 * sigmatilde
# muqK          = 1/qK * (qK_vartheta*vartheta_sigmatilde + qK_rhocheck*rhocheck_sigmatilde) *b*(sigmass-sigmatilde) + 0.5 * (qK_vartheta*vartheta_sigmatildesigmatilde + qK_rhocheck*rhocheck_sigmatildesigmatilde) *nu**2 * sigmatilde
################################################


varsigma      = -sigmav + vartheta*sigmaqB + (1-vartheta)*sigmaqK
varsigmatilde = gamma*(1-vartheta)*sigmatilde

rf_qB = muqB + np.log(phi*iota+1)/phi - delta - varsigma * sigmaqB
rf_qK = (a-iota)/qK + np.log(phi*iota+1)/phi - delta + muqK - varsigma * sigmaqK - varsigmatilde*sigmatilde

fig, ax = plt.subplots(2,3, figsize=(15,10))
ax[0,0].plot(sigmatilde, vartheta)
ax[0,0].set_ylabel(r'$\vartheta$')
ax[0,0].set_xlabel(r'$\tilde{\sigma}$')
ax[0,1].plot(sigmatilde, qK)
ax[0,1].set_ylabel(r'$q^K$')
ax[0,1].set_xlabel(r'$\tilde{\sigma}$')
ax[0,2].plot(sigmatilde, qB)
ax[0,2].set_ylabel(r'$q^B$')
ax[0,2].set_xlabel(r'$\tilde{\sigma}$')
ax[1,0].plot(sigmatilde, rf_qK, label=r'$r^{f}$ calculated by $q^K$')
ax[1,0].plot(sigmatilde, rf_qB, label=r'$r^{f}$ calculated by $q^\mathcal{B}$')
ax[1,0].set_ylabel(r'$r^{f}$')
ax[1,0].set_xlabel(r'$\tilde{\sigma}$')
ax[1,0].legend()
ax[1,1].plot(sigmatilde, varsigma)
ax[1,1].set_ylabel(r'$\varsigma$')
ax[1,1].set_xlabel(r'$\tilde{\sigma}$')
ax[1,2].plot(sigmatilde, varsigmatilde)
ax[1,2].set_ylabel(r'$\tilde{\varsigma}$')
ax[1,2].set_xlabel(r'$\tilde{\sigma}$')
plt.tight_layout()
plt.show()
```
